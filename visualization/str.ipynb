```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Example for loading and plotting a trace file
def plot_trace(file_path):
    df = pd.read_csv(file_path)
    
    # For each marker, plot alleles with their heights
    plt.figure(figsize=(12, 6))
    
    for _, row in df.iterrows():
        marker = row['Marker']
        alleles = row[['Allele 1', 'Allele 2', 'Allele 3']].dropna().values
        heights = row[['Height 1', 'Height 2', 'Height 3']].dropna().values
        
        for allele, height in zip(alleles, heights):
            plt.bar(allele, height, label=f"{marker}-{allele}")
    
    plt.title(f"STR Profile - {file_path}")
    plt.xlabel("Allele")
    plt.ylabel("Peak Height")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

```python
plot_trace("../datasets/NFI/Traces/1A2_New.csv")
```
